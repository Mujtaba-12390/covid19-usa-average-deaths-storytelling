```python
strt = 0
end = 3
deaths = []

for i in range(4):
    death_cumulative = death_toll["New_deaths"][strt:end].sum()
    deaths.append(death_cumulative)
    strt += 3
    end += 3
```

```python
deaths
```

```python
sum(deaths)
```

```python
for i in range(len(deaths)):
    if i > 0:
        deaths[i] = deaths[i-1] + deaths[i]
```

```python
deaths
```

```python
proportions = [round(death/max(deaths), 2) for death in deaths]
proportions
```

```python
xmax_vals = [round(0.5 + proportion * 0.3, 3) for proportion in proportions]
xmax_vals
```

```python
fig, (ax1,ax2,ax3,ax4) = plt.subplots(4, 1, figsize=(6,8))

axes = [ax1, ax2, ax3, ax4]

for ax in axes:
    ax.plot(death_toll["Month"], death_toll["New_deaths"], alpha = 0.2, color ="#b30000")
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    
    ax.tick_params(bottom = 0, left = 0)
    
    for location in ["left","right","bottom","top"]:
        ax.spines[location].set_visible(False)
    
    ax.text(0.5, -80, "0", alpha = 0.8)
    ax.text(12, 2400, format(2247,","), alpha = 0.8)
        
        
ax1.plot(death_toll["Month"][:3],death_toll["New_deaths"][:3], color ="#b30000", linewidth = 2.5)
ax2.plot(death_toll["Month"][2:6],death_toll["New_deaths"][2:6], color ="#b30000", linewidth = 2.5)
ax3.plot(death_toll["Month"][5:10],death_toll["New_deaths"][5:10], color ="#b30000", linewidth = 2.5)
ax4.plot(death_toll["Month"][9:12],death_toll["New_deaths"][9:12], color ="#b30000", linewidth = 2.5)


ax1.text(1.2, -300, "Jan - Mar", color ="#b30000", weight = "bold", rotation = 3)
ax2.text(3.7, 300, "Mar - Jun", color ="#b30000", weight = "bold",  rotation = 68)
ax3.text(7.2, 500, "Jul - Sep", color ="#b30000", weight = "bold",  rotation = 0)
ax4.text(10.5 , 600, "Oct - Dec", color ="#b30000", weight = "bold",  rotation = 45)

ax1.text(0.5, 3500, "The Virus Kills 908 People Each Month", size = 14, weight = "bold")
ax1.text(0.5, 3150, "Average Number of Month Deaths Per Month in the US", size = 12)


for ax in axes:
    ax.axhline(y=1600, xmin = 0.5, xmax = 0.8, color = "#b30000", linewidth = 6, alpha = 0.2)


for ax, xmax, death in zip(axes, xmax_vals, deaths):    
    ax.axhline(y=1600, xmin = 0.5, xmax = xmax , color = "#b30000", linewidth = 6)
    
    ax.text(7.5, 1850, format(death,","),  color = "#b30000", weight = "bold")

    
plt.savefig("storytelling.jpg")
plt.show()
```
